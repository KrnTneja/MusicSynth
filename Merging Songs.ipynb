```python
import scipy.io.wavfile as sciw
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sr1, y1 = sciw.read("Test_Bheegi.wav")
sr2, y2 = sciw.read("Test_Zalima.wav")
assert sr1 == sr2
y2 = y2[:,0][:len(y1)]
```

```python
addition = y1+y2
addition = np.int16(addition*1.0/np.max(np.absolute(addition))*(2**15-1))
plt.plot(addition)
plt.show()
```

```python
sciw.write(data=np.int16(y1+y2), filename="Test_Merge.wav", rate=sr1)
```

### Merging for Orchestra Effect

```python
sr, bass = sciw.read("bass.wav")
sr, bell = sciw.read("bell.wav")
sr, pads = sciw.read("pads.wav")
sr, compose1 = sciw.read("compose1.wav")
print len(bass)
print len(bell)
print len(pads)
print len(compose1)
```

```python
plt.plot(bass)
plt.show()
plt.plot(bell)
plt.show()
plt.plot(pads)
plt.show()
plt.plot(compose1)
plt.show()
```

```python
c = 2500 # Keep It Even
t = c*1.0/44100
print t
addition_orc = 0.2*bass[c/2:-c/2] + 0.4*bell[:-c] + 0.75*pads[c:]+0.6*compose1[c/2:-c/2]
addition_orc = np.int16(addition_orc*1.0/np.max(np.absolute(addition_orc))*(2**15-1))
plt.plot(addition_orc)
plt.show()
```

```python
sciw.write(data=np.int16(addition_orc), filename="Orch_5.wav", rate=sr)
```
